```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
matches_df = pd.read_csv("/content/IPL Matches 2008-2020.csv")
balls_df = pd.read_csv("/content/IPL Ball-by-Ball 2008-2020.csv")
```

```python
matches_df.head()
```

```python
matches_df['date'] = pd.to_datetime(matches_df['date'])
```

```python
matches_df['season'] = matches_df['date'].dt.year
```

```python
matches_df.head()
```

```python
balls_df
```

```python
#What was the count of matches played in each season?
matches_per_season = matches_df['season'].value_counts().sort_index()
matches_per_season
```

```python
#settings for the plots
sns.set_style("whitegrid")
plt.rcParams["figure.figsize"] = (10, 6)
```

```python
plt.figure()
sns.barplot(x=matches_per_season.index, y=matches_per_season.values, palette="viridis")
plt.xlabel('Season')
plt.ylabel('Number of Matches')
plt.title('Matches Played in Each Season')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

```python
#How many runs were scored in each season?
balls_df = pd.merge(balls_df, matches_df[['id', 'season']], on='id', how='left')
#merged_df = balls_df.merge(matches_df[['id', 'season']], on='id')
```

```python
runs_per_season = balls_df.groupby('season')['total_runs'].sum()
runs_per_season
```

```python
plt.figure()
sns.barplot(x=runs_per_season.index, y=runs_per_season.values, palette="magma")
plt.title('Runs Scored in Each Season')
plt.xlabel('Season')
plt.ylabel('Total Runs')
plt.xticks(rotation=45)
plt.tight_layout()
```

```python
#What were the runs scored per match in different seasons?
runs_per_match_season = balls_df.groupby(['id', 'season'])['total_runs'].sum()
runs_per_match_season
```

```python
#Who has umpired the most?
umpire = pd.concat([matches_df['umpire1'], matches_df['umpire2']])
umpire.value_counts().idxmax()
```

```python
#Which team has won the most tosses?
toss_winners = matches_df['toss_winner'].value_counts().idxmax()
toss_winners
```

```python
#What does the team decide after winning the toss?
toss_decision = matches_df['toss_decision'].value_counts()#.idxmax()
toss_decision
```

```python
plt.figure()
sns.barplot(x=toss_decision.index, y=toss_decision.values,hue = toss_decision.index, palette="magma")
plt.title('Toss Decision')
plt.xlabel('Decision')
plt.ylabel('Count')
```

```python
#How does the toss decision vary across seasons?
toss_decision_season = matches_df.groupby('season')['toss_decision'].value_counts().unstack()
toss_decision_season
```

```python
toss_decision_season.plot(kind='bar',stacked = True, colormap="Set2")
```

```python
#9. How many times has the chasing team won the match?
chasing_team_wins = matches_df[(matches_df['toss_decision'] == 'field') & (matches_df['toss_winner'] == matches_df['winner'])]
display(chasing_team_wins.shape[0])
```

```python
#10. Which all teams had won this tournament?
last_match_per_season = matches_df.groupby('season')['winner'].tail(1)
last_match_per_season
```

```python
#11. Which team has played the most number of matches?
matches_per_team = pd.concat([matches_df['team1'], matches_df['team2']])
matches_per_team.value_counts().idxmax()
```

```python
#12. Which team has won the most number of times?
matches_df['winner'].value_counts().idxmax()
```

```python
#13. Which team has the highest winning percentage?
winning_percentage = matches_df['winner'].value_counts() / matches_df['winner'].count() * 100
winning_percentage.idxmax()
```

```python
#14. Is there any lucky venue for a particular team?
lucky_venue = matches_df.groupby('venue')['winner']
lucky_venue.value_counts().idxmax()
```

```python
# 15. Innings wise comparison between teams
innings_comparison = balls_df.groupby(['inning', 'batting_team'])['total_runs'].sum()
innings_comparison
```

```python
#16. Which team has scored the most number of 200+ scores?
most_score = balls_df.groupby('batting_team')['total_runs'].sum() >= 200
most_score.idxmax()
```

```python
#17. Which team has conceded 200+ scores the most?
most_conceded = balls_df.groupby('bowling_team')['total_runs'].sum() >= 200
most_conceded.idxmax()
```

```python
#18. What was the highest run scored by a team in a single match?
highest_run = balls_df.groupby(['id', 'inning'])['total_runs'].sum().max()
highest_run
```

```python
#19. Which is the biggest win in terms of run margin?
biggest_win = matches_df.groupby('winner')['result_margin'].max()
biggest_win.idxmax()
```

```python
#19. Which is the biggest win in terms of run margin?
biggest_win = matches_df.loc[matches_df['result_margin'].idxmax(), ['winner', 'result_margin']]
display(biggest_win)
```

```python
#20. Which batsmen have played the most number of balls?
batsman_most_score = balls_df.groupby('batsman')['ball'].sum().nlargest(1)
batsman_most_score


#batsman_most_score = balls_df.groupby('batsman')['over'].sum().nlargest(1)
#batsman_most_score
```

```python
#21. Who are the leading run-scorers of all time?
top_run_scorers = balls_df.groupby('batsman')['batsman_runs'].sum().nlargest(5)
top_run_scorers
```

```python
#22. Who has hit the most number of 4's?
most_4 = balls_df[balls_df['batsman_runs'] == 4].groupby('batsman')['batsman_runs'].count().nlargest(1)
most_4
```

```python
#23. Who has hit the most number of 6's?
most_6 = balls_df[balls_df['batsman_runs'] == 6].groupby('batsman')['batsman_runs'].count().nlargest(1)
most_6
```

```python
#24. Who has the highest strike rate?
strike_rate = balls_df.groupby('batsman')['batsman_runs'].sum() / balls_df.groupby('batsman')['ball'].sum() * 100
strike_rate.nlargest(1)
```

```python
#25. Who is the leading wicket-taker?
wicket_taker = balls_df.groupby('bowler')['is_wicket'].sum()
wicket_taker.nlargest(1)
```

```python
#26. Which stadium has hosted the most number of matches?
most_hosted_stadium = matches_df['venue'].value_counts()
most_hosted_stadium.idxmax()
```

```python
#27. Who has won the most MOM awards?
man_of_match = matches_df['player_of_match'].value_counts()
man_of_match.idxmax()
```

```python
#28. What is the count of fours hit in each season?
fours_per_season = balls_df[balls_df['batsman_runs'] == 4].groupby('season')['batsman_runs'].count()
fours_per_season
```

```python
#29. What is the count of sixes hit in each season?
six_per_season = balls_df[balls_df['batsman_runs'] == 6].groupby('season')['batsman_runs'].count()
six_per_season
```

```python
#Visualiztion
plt.figure(figsize=(9, 5))
sns.barplot(x=six_per_season.head(10).index, y=six_per_season.head(10).values, palette='viridis')
plt.xlabel('Season')
plt.ylabel('Number of Sixes')
plt.title('Count of Sixes Hit in Each Season')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

```python
#30. What is the count of runs scored from boundaries in each season?
boundaries_per_season = balls_df[(balls_df['batsman_runs'] == 4) | (balls_df['batsman_runs'] == 6)].groupby('season')['batsman_runs'].sum()
boundaries_per_season
```

```python
#Visualization
plt.figure(figsize=(9, 5))
sns.barplot(x=boundaries_per_season.head(10).index, y=boundaries_per_season.head(10).values, palette='viridis')
plt.xlabel('Season')
plt.ylabel('Number of Runs')
plt.title('Count of Runs Scored from Boundaries in Each Season')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

```python
# 31. What is the run contribution from boundaries in each season?
run_contribution_boundaries = boundaries_per_season / runs_per_season * 100
run_contribution_boundaries
```

```python
# 32. Which team has scored the most runs in the first 6 overs?
first_six_overs_runs = balls_df[balls_df['over'] <= 6].groupby('batting_team')['total_runs'].sum().nlargest(1)
first_six_overs_runs
```

```python
# 33. Which team has scored the most runs in last 4 overs?
last_four_overs_runs = balls_df[balls_df['over'] >= 16].groupby('batting_team')['total_runs'].sum().nlargest(1)
last_four_overs_runs
```

```python
# 34. Which team has the best scoring run-rate in the first 6 overs?
run_rate_first_6_overs = balls_df[balls_df['over'] <= 6].groupby('batting_team')['total_runs'].sum() / 6
run_rate_first_6_overs.idxmax()
```

```python
# 35. Which team has the best scoring run-rate in  the last 4 overs?
last_4_overs_run_rate = balls_df[balls_df['over'] >= 16].groupby('batting_team')['total_runs'].sum() / 4
last_4_overs_run_rate.idxmax()
```
